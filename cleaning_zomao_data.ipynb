# Removing ',' from average_cost

```python
def clean_cost(x):
    if isinstance(x, str):
        return(x.replace(',', ''))
    return(x)

dataset['average_cost']=dataset['average_cost'].apply(clean_cost).astype('float')
```

```python
dataset['average_cost'].value_counts()
```

# Dish data

```python
dishes_data = dataset[dataset.dish_liked.notnull()]
dishes_data.dish_liked = dishes_data.dish_liked.apply(lambda x:x.lower().strip())
```

Missing Dish Data

```python
dishes_data.isnull().sum()
```

# Getting Most Liked Dish

```python
dish_count = []
for i in dishes_data.dish_liked:
    for t in i.split(','):
        t = t.strip() # remove the white spaces to get accurate results
        dish_count.append(t)
```

```python
plt.figure(figsize=(12,6)) 
pd.Series(dish_count).value_counts()[:10].plot(kind='bar',color= 'g')
plt.title('Top 10 dished_liked in Bangalore',weight='bold')
plt.xlabel('Dish')
plt.ylabel('Count')
```

# Ratings

```python
dataset['rates'].isnull().sum()
```

Taking Care Of Missing Rates

```python
dataset['rates'].isnull()
dataset['rates'] = dataset['rates'].replace('NEW',np.NaN)
dataset['rates'] = dataset['rates'].replace('-',np.NaN)
dataset.dropna(how = 'any', inplace = True)
```

```python
dataset['rates'] = dataset.loc[:,'rates'].replace('[ ]','',regex = True)
dataset['rates'] = dataset['rates'].astype(str)
dataset['rates'] = dataset['rates'].apply(lambda r: r.replace('/5',''))
dataset['rates'] = dataset['rates'].apply(lambda r: float(r))
```

```python
dataset.rates.hist(color='green')
plt.axvline(x= dataset.rates.mean(),ls='--',color='yellow')
plt.title('Average Rating for Bangalore Restaurants',weight='bold')
plt.xlabel('Rating')
plt.ylabel('No of Restaurants')
print(dataset.rates.mean())
```

# Types of Restaurants VS Ratings

```python
f,ax=plt.subplots(figsize=(18,8))
g = sns.pointplot(x=dataset["rest_type"], y=dataset["rates"], data=dataset)
g.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.title('Restaurent type vs Rate', weight = 'bold')
plt.show()
```

# Cuisines

```python
cuisines_data = dataset[dataset.cuisines.notnull()]
cuisines_data.cuisines = cuisines_data.cuisines.apply(lambda x:x.lower().strip())
```

# Top 10 Cuisines

```python
cuisines_count= []
for i in cuisines_data.cuisines:
    for j in i.split(','):
        j = j.strip()
        cuisines_count.append(j)
```

```python
plt.figure(figsize=(12,6)) 
pd.Series(cuisines_count).value_counts()[:10].plot(kind='bar',color= 'y')
plt.title('Top 10 cuisines in Bangalore',weight='bold')
plt.xlabel('cuisines type')
plt.ylabel('No of restaurants')
```
